```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data - Replace this with your actual dataset
data = {
    'CustomerID': range(1, 21),
    'LoyaltyScore': ['High', 'Medium', 'Low', 'High', 'Medium', 'Low', 'High', 'High', 'Medium', 'Low',
                     'Medium', 'High', 'Low', 'Medium', 'High', 'Low', 'Medium', 'High', 'Low', 'High'],
    'EngagementStatus': ['Active', 'Dormant', 'Churned', 'Active', 'Dormant', 'Churned', 'Active', 'Dormant', 'Churned', 'Active',
                         'Dormant', 'Churned', 'Active', 'Dormant', 'Churned', 'Active', 'Dormant', 'Churned', 'Active', 'Dormant']
}

df = pd.DataFrame(data)

# Set style
sns.set(style="whitegrid")

# Loyalty Score Distribution (Bar Plot)
plt.figure(figsize=(8, 5))
sns.countplot(x='LoyaltyScore', data=df, order=['High', 'Medium', 'Low'], palette='pastel')
plt.title('Loyalty Score Distribution')
plt.xlabel('Loyalty Score')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()

# Engagement Status Distribution (Pie Chart)
engagement_counts = df['EngagementStatus'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(engagement_counts, labels=engagement_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Customer Engagement Status')
plt.tight_layout()
plt.show()
```
